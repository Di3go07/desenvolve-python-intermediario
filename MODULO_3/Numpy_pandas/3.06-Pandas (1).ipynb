#### Q4. 
Crie um dataframe `participacao` com as colunas `país`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada país participou, informando se é da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 países de cada competição que mais participou de copas do mundo.

Exemplo de saída (valores inventados):
```
país            copa        num_copas
Brazil          Feminina    8
Unites States   Feminina    8
Germany         Feminina    8
Japan           Feminina    7

```

```
país            copa        num_copas
Brazil          Masculina   21
Germany         Masculina   21
Argentina       Masculina   20
England         Masculina   20
Mexico          Masculina   20
```

````python
import pandas as pd

#DATAFRAME
participacao = {
    "pais": [],
    "copa": [],
    "num_copas": []
}


#FILTRAGEM
#lista com todos os países que já participaram
lista_paises = []

nomes_paises = wc["time_1"].drop_duplicates() #buscando a coluna com o nome dos países sem repeti-los

for linha in nomes_paises:
    lista_paises.append(linha)
    
###print(lista_paises)

#resgatar quantas participações o país teve
wc_filtrado = wc[["time_1", "ano", "copa"]].drop_duplicates() #busca o nome dos países participantes em cada ano

#PROCESSAMENTO
#looping para adicionar ao dataframe informações de cada país que já participou da copa
for pais in lista_paises:
    #Masculino
    participacoes_pais = (wc_filtrado[(wc_filtrado["time_1"] == pais) & (wc_filtrado["copa"] == "Masculina")])
    ###print(f"{pais}: {len(participacoes_pais)}")
    if len(participacoes_pais) > 0: #alguns países só participaram da copa Feminina
        participacao['pais'].append(pais)
        participacao['copa'].append("Masculina")
        participacao['num_copas'].append(len(participacoes_pais)) 
        
    #Feminino
    participacoes_pais = (wc_filtrado[(wc_filtrado["time_1"] == pais) & (wc_filtrado["copa"] == "Feminina")])
    ###print(f"{pais}: {len(participacoes_pais)}")
    if len(participacoes_pais) > 0: #alguns países só participaram da copa Masculina
        participacao['pais'].append(pais)
        participacao['copa'].append("Feminina")
        participacao['num_copas'].append(len(participacoes_pais)) 

#RESULTADO 
df = pd.DataFrame(participacao) #database completa

#TOP 5 FEMININO
top5_fem = df[df["copa"] == 'Feminina'].sort_values(by='num_copas', ascending=False).head(5).reset_index(drop=True)
top5_fem.index = top5_fem.index + 1 #para começar a contar do 1
display(top5_fem)

#TOP 5 MASCULINO
top5_mas = df[df["copa"] == 'Masculina'].sort_values(by='num_copas', ascending=False).head(5).reset_index(drop=True)
top5_mas.index = top5_mas.index + 1
display(top5_mas)
````

#### Q6. 
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista. 

Lembre de levar em consideração que muitas observações são `NaN`. 

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
Cartoes_Amarelos = {
    "pais": [],
    "cartões amarelos": []
}

#FILTRAGEM
#lista com todos os países que já participaram
lista_paises = []

nomes_paises = wc["time_1"].drop_duplicates() #buscando a coluna com o nome dos países sem repeti-los

for linha in nomes_paises:
    lista_paises.append(linha)

#jogos em que os times levaram cartão
cartoes_1 = wc[(wc["cartao_amarelo_1"].notna())] #como time 1
cartoes_2 = wc[(wc["cartao_amarelo_2"].notna())] #como time 2

#PROCESSAMENTO

#iterando sobre cada país
for pais in lista_paises:
    ###AMARELO
    #TABELA CARTÃO 1
    #quais jogos o time levou cartão
    cartoes_pais_1 = cartoes_1[(cartoes_1["time_1"] == pais)][["cartao_amarelo_1", "time_1"]].reset_index(drop=True)

    #tratamento de dados
    cartoes_pais_1["cartao_amarelo_1"] = cartoes_pais_1["cartao_amarelo_1"].fillna("").apply(lambda x: x.split(","))

    #contando o numero de cartão por jogo
    cartoes_pais_1["quantidade_cartao"] = cartoes_pais_1["cartao_amarelo_1"].apply(len)

    #TABELA CARTÃO 2
    cartoes_pais_2 = cartoes_2[(cartoes_2["time_2"] ==  pais)][["cartao_amarelo_2", "time_2"]].reset_index(drop=True)
    cartoes_pais_2["cartao_amarelo_2"] = cartoes_pais_2["cartao_amarelo_2"].fillna("").apply(lambda x: x.split(","))
    cartoes_pais_2["quantidade_cartao"] = cartoes_pais_2["cartao_amarelo_2"].apply(len)

    #Operacao
    #somando a quantidade de cartões que o time levou
    soma_cartoes1 = cartoes_pais_1["quantidade_cartao"].sum() #variável com a soma dos valores da coluna
    soma_cartoes2 = cartoes_pais_2["quantidade_cartao"].sum()
    #calculando o total
    soma_total = soma_cartoes1 + soma_cartoes2
    
    #Adicionando ao novo DataFrame
    Cartoes_Amarelos['pais'].append(pais)
    #quantidade de cartões que o país já levou
    Cartoes_Amarelos['cartões amarelos'].append(soma_total)

#RESULTADO 
df = pd.DataFrame(Cartoes_Amarelos) #database completa

display(df.sort_values(by='cartões amarelos', ascending=False).reset_index(drop=True))
````

#### Q7.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`. 

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
Jogadores = {
    "jogador": [],
    "num_gols": []
}

jogador_por_gol = []

#separando as tabelas
gols_1 = wc[(wc["gols_1_detalhes"].notna())] #como time 1
gols_2 = wc[(wc["gols_2_detalhes"].notna())] #como time 2
penalti_1 = wc[(wc["gols_1_penalti"].notna())] 
penalti_2 = wc[(wc["gols_2_penalti"].notna())] 

#tratamento de dados
#GOLS 1
gols_1.loc[:, ('gols_1_detalhes')] = gols_1["gols_1_detalhes"].apply(lambda x: x.split("|"))
#separando cada gol

#looping para retirar o nome de cada jogador na lista em cada linha
for index, row in gols_1.iterrows():
    for item in range(len(row['gols_1_detalhes'])):
        nome_jogador = row['gols_1_detalhes'][item].split('·')
        jogador_por_gol.append(nome_jogador[0].strip()) #remover o espaço em branco no final da string

#GOLS 2
gols_2.loc[:, ('gols_2_detalhes')] = gols_2["gols_2_detalhes"].apply(lambda x: x.split("|"))

for index, row in gols_2.iterrows():
    for item in range(len(row['gols_2_detalhes'])):
        nome_jogador = row['gols_2_detalhes'][item].split('·')
        jogador_por_gol.append(nome_jogador[0].strip()) #remover o espaço em branco no final da string
        
#PENALTI 1        
penalti_1.loc[:, ('gols_1_penalti')] = penalti_1["gols_1_penalti"].apply(lambda x: x.split("|"))

for index, row in penalti_1.iterrows():
    for item in range(len(row['gols_1_penalti'])):
        linha_jogador = row['gols_1_penalti'][item].split('·')
        nome_jogador = linha_jogador[0]       
        nome_corrigido = nome_jogador.replace(' (P)', '') #remover o (P) após o nome
        jogador_por_gol.append(nome_corrigido.strip()) #remover o espaço em branco no final da string

#PENALTI 2       
penalti_2.loc[:, ('gols_2_penalti')] = penalti_2["gols_2_penalti"].apply(lambda x: x.split("|"))

for index, row in penalti_2.iterrows():
    for item in range(len(row['gols_2_penalti'])):
        linha_jogador = row['gols_2_penalti'][item].split('·')
        nome_jogador = linha_jogador[0]       
        nome_corrigido = nome_jogador.replace(' (P)', '') 
        jogador_por_gol.append(nome_corrigido.strip()) #remover o espaço em branco no final da string
lista_jogadores = list(set(jogador_por_gol))

#adicionando valor ao dataframe
for jogador in lista_jogadores:
    Jogadores['jogador'].append(jogador)
    num_gols = jogador_por_gol.count(jogador)
    Jogadores['num_gols'].append(num_gols)

#RESULTADO 
df = pd.DataFrame(Jogadores) #database completa
display(df.sort_values(by='num_gols', ascending=False).reset_index(drop=True))
````
